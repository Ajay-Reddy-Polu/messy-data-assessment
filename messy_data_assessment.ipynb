```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'customer_id':[1,2,3,4,2,5],
    'age':[25,np.nan,35,np.nan,np.nan,40],
    'city':['Mumbai','Delhi',np.nan,'Mumbai','Delhi',np.nan],
    'gender':['Male','Female','Female','Male','Female','Male'],
    'annual_income':[40000,80000,60000,120000,80000,95000]
})

```

Impute missing values in age with the median and in city with the mode

```python
df.fillna({
    'age':df['age'].median(),
    'city':df['city'].mode()[0]
}, inplace=True)
```

Remove any duplicate rows

```python
df = df.drop_duplicates()
```

```python
df
```

Task 2 — Encode Categorical Columns
After cleaning

 encode the categorical columns:

Apply One-Hot Encoding to city

Apply Label Encoding to gender

Print the resulting dataframe.

```python
from sklearn.preprocessing import LabelEncoder

#one-hot Encoding from city
df = pd.get_dummies(df, columns=['city'])

#Label Encoding for gender
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])

print("\nEncoded Data:\n",df)
```

Task 3 — Scale Numerical Features

Scale age and annual_income using both methods below on the cleaned dataframe,
 and print the results of each:

Min-Max Scaling (using MinMaxScaler)


```python
from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler()
df_minmax = df.copy()

df_minmax[['age','annual_income']] = minmax.fit_transform(
    df_minmax[['age','annual_income']]
)

print("\nMinMax Scaled Data:\n",df_minmax)
```

Standardisation (using StandardScaler)

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_standard = df.copy()

df_standard[['age','annual_income']] = scaler.fit_transform(
    df_standard[['age','annual_income']]
)

print("\nStandardized Data:\n",df_standard)
```

MIN - Max Scaling:

converts values between 0 and 1

best when data has no outliers

used in models like KNN,Neural Networks

STANDARDIZATION:

converts data to mean = 0  and std = 1

handles outliers better

used in LINEAR REGRESSION ,SVM
